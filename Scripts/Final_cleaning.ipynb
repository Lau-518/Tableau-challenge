```python
#import dependencies

import pandas as pd
import numpy as np
```

```python
# Load in file
file_path = "2019-09_2020-06.csv"
```

```python
# Read and display the CSV with Pandas
summary_df = pd.read_csv(file_path,encoding='ISO-8859-1')
summary_df.head(5)
```

```python
#Deleting members that are over 65 years old (born before 1956)

summary_df.drop(summary_df[summary_df['birth year'] < 1956].index, inplace = True)
```

```python
summary_df['birth year'].min()
```

```python
#Replacing Gender (0 = unknown & 1 = male & 2 = female)
summary_df.loc[(summary_df.gender == 0),'gender']='unknown'
summary_df.loc[(summary_df.gender == 1),'gender']='male'
summary_df.loc[(summary_df.gender == 2),'gender']='female'

```

```python
summary_df.head()
```

```python
#Exporting cleaned dataframe into csv
summary_df.to_csv('Cleaned_data_2019-09_2020-06.csv', index=False)
```
